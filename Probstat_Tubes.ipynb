$$ 
t_0 = \frac{\bar{X}-\mu_0}{s/\sqrt{n}}

akan dihitung nilai $p_{value}$

```python
pVal = 1-t.cdf(t0,  df["pH"].size - 1)
pVal
```

### 5.a. Data kolom fixed acidity dibagi 2 sama rata: bagian awal dan bagian akhir kolom. Benarkah rata-rata kedua bagian tersebut sama

```python
dataAwal = df["fixed acidity"].iloc[:df["fixed acidity"].size//2]
dataAkhir = df["fixed acidity"].iloc[df["fixed acidity"].size//2:]
```

```python

sns.boxplot(x = dataAwal)
```

```python

sns.boxplot(x = dataAkhir)
```

#### Langkah 1: Penentuan $H_0$

$$
H_0: \mu_{awal} - \mu_{akhir} = 0
$$

#### Langkah 2: Penentuan $H_1$

$$
H_1: \mu_{awal} - \mu_{akhir} \neq 0
$$

#### Langkah 3: Penentuan $\alpha$

$$ \alpha = 0.05 $$

#### Langkah 4: Pengujian Statistik dan Daerah Kritis

$$
z = \frac{(\bar{x}_1 - \bar{x}_2) - (\mu_{1}-\mu_{2})}{\sqrt{\sigma_1^2/n_1 + \sigma_2^2/n_2}}
$$

Daerah kritisnya adalah 
$$ z<-z_{\alpha/2} \lor z>z_{\alpha/2} $$

#### Langkah 5: Hitung nilai uji statistik dari data sample. Hitung p-value sesuai dengan uji statistik yang digunakan.

```python
x1bar = dataAwal.mean()
x2bar = dataAkhir.mean()
sigma1 = dataAwal.std()
sigma2 = dataAkhir.std()
n1 = dataAwal.count()
n2 = dataAkhir.count()

z = ((x1bar-x2bar)-0)/np.sqrt(sigma1**2/n1+sigma2**2/n2)

pVal = 2*(1-norm.cdf(z))

zaper2 = norm.ppf(1-0.05/2)
print(f"z = {z}")
print(f"p = {pVal}")
print(f"z_a/2 = {zaper2}")
```

#### Langkah 6: Pengambilan Keputusan

$H_0$ tidak ditolak karena $ |z| < 1.9599 $ <br>
$H_0$ tidak ditolak semakin kuat karena $p_{value} \ge 0.05$

#### Langkah 4: Pengujian Statistik dan Daerah Kritis

$$
f = s_1^2/s_2^2

$$ f>f_{1-\alpha/2}(v_1,v_2) \lor f<f_{\alpha/2}(v_1,v_2) $$

### 5.e. Proporsi nilai setengah bagian awal alcohol yang lebih dari 7, adalah lebih besar daripada, proporsi nilai yang sama di setengah bagian akhir alcoho

```python
dataAwal = df["alcohol"].iloc[:df["alcohol"].size//2]
dataAkhir = df["alcohol"].iloc[df["alcohol"].size//2:]
```

```python

sns.boxplot(x = dataAwal)
```

```python

sns.boxplot(x = dataAkhir)
```

#### Langkah 1: Penentuan $H_0$

$$
H_0: p_{awal} - p_{akhir} = 0
$$

#### Langkah 2: Penentuan $H_1$

$$
H_1: p_{awal} - p_{akhir} > 0
$$

#### Langkah 3: Penentuan $\alpha$

$$ \alpha = 0.05 $$

#### Langkah 4: Pengujian Statistik dan Daerah Kritis

$$
z = \frac{\hat{p}_1 - \hat{p}_2}{\sqrt{\hat{p}\hat{q}(1/n_1+1/n_2)}}
$$

Daerah kritisnya adalah 
$$ z>z_\alpha $$

#### Langkah 5: Hitung nilai uji statistik dari data sample. Hitung p-value sesuai dengan uji statistik yang digunakan.

```python
x1 = dataAwal[dataAwal>7].count()
x2 = dataAkhir[dataAkhir>7].count()
n1 = dataAwal.count()
n2 = dataAkhir.count()

p = (x1+x2)/(n1+n2)
q = 1-p

p1 = x1/n1
p2 = x2/n2

z = (p1 - p2)/(np.sqrt(p*q*(1/n1+1/n2)))

pVal = 1-norm.cdf(z)

z_a = abs(norm.ppf(0.05))
print(f"z = {z}")
print(f"p = {pVal}")
print(f"z_a = {z_a}")
```

#### Langkah 6: Pengambilan Keputusan

$H_0$ tidak ditolak karena $ z < 1.644 $
