```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis, shapiro, kstest
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("anggur.csv")
df
```

```python
print(df.dtypes)
```

```python
mean = df.mean()
mean
```

Median

```python
median = df.median()
median
```

```python
mode = df.mode().iloc[0]
mode
```

```python
std = df.std()
std
```

```python
var = df.var()
var
```

```python
minimum = df.min()
minimum
```

```python
maximum = df.max()
maximum
```

```python
q1 = df.quantile(0.25)
q1
```

```python
q2 = df.quantile(0.5)
q2
```

```python
q3 = df.quantile(0.75)
q3
```

```python
iqr = q3 - q1
iqr
```

```python
skewness = df.apply(lambda x: skew(x.dropna()))
skewness
```

```python
kurt = df.apply(lambda x: kurtosis(x.dropna()))
kurt
```

```python
df_summary = pd.DataFrame({'mean': mean, 'median': median, 'mode': mode, 'std': std, 'var': var, 'min': minimum, 'max': maximum, 'Q1': q1, 'Q2': q2, 'Q3': q3, 'IQR': iqr, 'skewness': skewness, 'kurtosis': kurt})
```

```python
df_summary
```

```python
for col in df.select_dtypes(include=[np.number]):
    fig, axs = plt.subplots(ncols=2, figsize=(10, 5))
    sns.histplot(df[col], ax=axs[0], kde=True)
    sns.boxplot(df[col], ax=axs[1])
    fig.suptitle(f'Distribusi {col}', fontsize=14)
    plt.show()
```

```python
# menentukan apakah setiap kolom numerik berdistribusi normal dengan menggunakan normality test Shapiro-Wilk dan Kolmogorov-Smirnov
for col in df.select_dtypes(include=[np.number]):
    print(f'Kolom {col}:')
    # normality test Shapiro-Wilk
    stat, p = shapiro(df[col])
    alpha = 0.05
    if p > alpha:
        print(f'    Shapiro-Wilk: p-value = {p:.4f}, distribusi normal (tidak ditolak H0)')
    else:
        print(f'    Shapiro-Wilk: p-value = {p:.4f}, distribusi tidak normal (ditolak H0)')
#     # normality test Kolmogorov-Smirnov
#     stat, p = kstest(df[col], 'norm')
#     if p > alpha:
#         print(f'    Kolmogorov-Smirnov: p-value = {p:.4f}, distribusi normal (tidak ditolak H0)')
#     else:
#         print(f'    Kolmogorov-Smirnov: p-value = {p:.4f}, distribusi tidak normal (ditolak H0)')
    # visualisasi histogram plot
    plt.figure(figsize=(5, 4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Histogram {col}')
    plt.show()
```
